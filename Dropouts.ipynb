```python
x = df.drop(60,axis='columns')
y = df[60]
```

```python
y = pd.get_dummies(y,drop_first=True)
y
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state = 1)
```

```python
df.shape
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dense(30, activation='relu'), #30,15 for trail and error
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')#binary classifer so 1
])
model.compile(loss='binary_crossentropy',optimizer ='adam', metrics=['accuracy'])#binary_crossentrophy this is used for calculate the loss --- it is binaryclassification so ..this binaryent
model.fit(X_train, y_train, epochs = 80, batch_size = 8)
```

```python
model.evaluate(X_test,y_test)#76 percent of dataset only good balance 20 percent in dataset were not predicated well
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])
```

```python
#for rounding the above predicated values
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
#comparing the test and predicated value
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix , classification_report
print(classification_report(y_pred,y_test))
```

keras dropout layer
it drops randomly

```python
modeld = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.5),#it drops 50 percent of neurons
    keras.layers.Dense(30, activation='relu'), #30,15 for trail and error
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')#binary classifer so 1

])
modeld.compile(loss='binary_crossentropy',optimizer ='adam', metrics=['accuracy'])#binary_crossentrophy this is used for calculate the loss --- it is binaryclassification so ..this binaryent
modeld.fit(X_train, y_train, epochs = 90, batch_size = 8)
```

```python
modeld.evaluate(X_test,y_test)#accuracy was increased after dropout some neurons
```

```python
y_pred = modeld.predict(X_test).reshape(-1)
print(y_pred[:10])
```

```python
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
from sklearn.metrics import confusion_matrix , classification_report
print(classification_report(y_pred,y_test))
```

```python
y_pred = modeld.predict(X_test).reshape(-1)
y_pred = np.round(y_pred)


from sklearn.metrics import confusion_matrix , classification_report
print(classification_report(y_pred,y_test))
```
